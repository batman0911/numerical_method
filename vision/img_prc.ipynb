```python
import numpy as np
import matplotlib.pylab as plt
```

```python
img = plt.imread("../data/Lenna.png")
```

```python
im = np.zeros(img.shape, dtype="uint8")
im = img * 256
im = im.astype("uint8")
```

```python
print(im)
print(img.shape)
```

```python
rgb_weights = np.array([0.333, 0.333, 0.333])
grayscale_img = np.dot(img[..., :3], rgb_weights)
```

```python
grayscale_img.shape
```

```python
plt.imshow(grayscale_img)
```

rotate image

```python
img_90_1 = np.rot90(img, 1)
img_90_2 = np.rot90(img, 2)
img_90_3 = np.rot90(img, 3)
```

```python
plt.imshow(img_90_1)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

# axs[0].imshow(img)

for c, ax in zip(range(3), axs):
    tmp_im = np.zeros(im.shape, dtype="uint8")
    tmp_im[:,:,c] = im[:,:,c]
    # print(tmp_im)
    ax.imshow(tmp_im)
    ax.set_axis_off()
```
